# **AI Chatbot for Customer Support**

```
# Importing required libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import gradio as gr

# Load pretrained model and tokenizer
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Chat history
chat_history = []

def chatbot(user_input):
    global chat_history
    # Tokenize the input and add it to the chat history
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')
    bot_input_ids = new_input_ids if len(chat_history) == 0 else torch.cat([chat_history[-1], new_input_ids], dim=-1)

    # Generate response
    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    chat_history.append(chat_history_ids)

    # Decode the response
    response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
    return response

# Create Gradio GUI
iface = gr.Interface(
    fn=chatbot,
    inputs="text",
    outputs="text",
    title="AI Chatbot for Customer Support",
    description="Type your queries to get a response from the AI chatbot."
)

# Launch the interface
iface.launch()
```

# **Object Detction with Pretrained Model**


```
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("object-detection", model="facebook/detr-resnet-50")
```

```
# Load model directly
from transformers import AutoImageProcessor, AutoModelForObjectDetection

processor = AutoImageProcessor.from_pretrained("facebook/detr-resnet-50")
model = AutoModelForObjectDetection.from_pretrained("facebook/detr-resnet-50")
```

```
# prompt: now create some demo to use above model "facebook/detr-resnet-50". Show Some examples to upload and Perform Image segmentation

import torch
import numpy as np
from PIL import Image
import requests
from io import BytesIO
from google.colab import files
import gradio as gr

def predict(image):
    # Ensure the input is a PIL Image
    if isinstance(image, np.ndarray):
        image = Image.fromarray(image)

    # Perform object detection
    inputs = processor(images=image, return_tensors="pt")
    outputs = model(**inputs)

    # convert outputs (bounding boxes and class logits) to COCO API
    # let's only keep detections with score > 0.9
    target_sizes = torch.tensor([image.size[::-1]])
    results = processor.post_process_object_detection(outputs, target_sizes=target_sizes, threshold=0.9)[0]

    # Draw bounding boxes on the image (simple visualization)
    img_with_boxes = image.copy()
    from PIL import ImageDraw
    draw = ImageDraw.Draw(img_with_boxes)

    detections = []
    for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
        box = [round(i, 2) for i in box.tolist()]
        label_text = model.config.id2label[label.item()]
        detections.append(f"Detected {label_text} with confidence {round(score.item(), 3)} at location {box}")

        # Draw bounding box
        draw.rectangle(box, outline="red", width=2)
        # Optionally add label text
        # draw.text((box[0], box[1]), label_text, fill="red")


    return img_with_boxes, "\n".join(detections)

# Create Gradio interface
iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil"),
    outputs=[gr.Image(label="Detected Objects"), gr.Textbox(label="Detection Results")],
    title="Image Object Detection with DETR",
    description="Upload an image to detect objects using the facebook/detr-resnet-50 model."
)

# Launch the interface
iface.launch(debug=True)
```

```

import gradio as gr
from transformers import AutoImageProcessor, SegformerForSemanticSegmentation
from PIL import Image
import torch

# Load the image processor and model
# This downloads the necessary components for the Segformer model.
processor = AutoImageProcessor.from_pretrained("nvidia/segformer-b0-finetuned-ade-512-512")
model = SegformerForSemanticSegmentation.from_pretrained("nvidia/segformer-b0-finetuned-ade-512-512")

def segment_uploaded_image(image: Image.Image):
    """
    Performs semantic segmentation on a PIL Image uploaded by the user.

    Args:
        image: A PIL Image object provided by Gradio's uploader.

    Returns:
        A PIL Image with the segmentation mask overlaid.
    """
    # Ensure the image is in RGB format, necessary for the model
    if image.mode != "RGB":
        image = image.convert("RGB")

    # Prepare the image for the model
    # This involves resizing and normalizing the image according to the model's requirements.
    inputs = processor(images=image, return_tensors="pt")

    # Perform inference (segmentation) using the model
    # torch.no_grad() is used because we are not training, saving memory and speed.
    with torch.no_grad():
        outputs = model(**inputs)

    # Process the model's output
    # The logits are the raw output scores for each class at each pixel.
    logits = outputs.logits.cpu() # Move to CPU for processing if using GPU

    # Rescale the logits to the original image size
    # The model produces smaller output maps, so we upsample them.
    upsampled_logits = torch.nn.functional.interpolate(
        logits,
        size=image.size[::-1], # Interpolate to (height, width) of original image
        mode="bilinear",        # Use bilinear interpolation
        align_corners=False,    # Align corners for consistent results
    )

    # Get the predicted segmentation mask
    # For each pixel, the class with the highest logit is the predicted segment.
    seg = upsampled_logits.argmax(dim=1)[0] # Get the mask for the first image in the batch

    # Create a visualization of the segmentation mask
    # Convert the mask to a PIL Image in indexed color mode (P)
    seg_img = Image.fromarray(seg.byte().numpy()).convert("P")
    # Convert to RGB to allow blending with the original image
    seg_img = seg_img.convert("RGB")

    # Overlay the segmentation mask on the original image
    # Image.blend combines the original image and the mask with a certain alpha transparency.
    # This makes the segmentation visible on top of the original image.
    overlay_img = Image.blend(image.convert("RGBA"), seg_img.convert("RGBA"), alpha=0.5) # Adjust alpha for desired transparency

    return overlay_img

# Set up the Gradio Interface
# This creates a user-friendly web interface.
iface_segmentation = gr.Interface(
    fn=segment_uploaded_image,  # The function to run when the user interacts
    inputs=gr.Image(type="pil", label="Upload Image"), # Input component: an image uploader, returns a PIL Image
    outputs=gr.Image(label="Segmentation Mask Overlay"), # Output component: displays a PIL Image
    title="Image Segmentation with Segformer", # Title for the interface
    description="Upload an image to see its semantic segmentation using the nvidia/segformer-b0-finetuned-ade-512-512 model." # Description
)

# Launch the Gradio interface
# This starts the web server for the interface.
# debug=True provides more detailed output in the console.
iface_segmentation.launch(debug=True)

# Note: GradientAccumulator is a training technique and is not relevant for inference (just running predictions).
# The code above focuses on the simple inference task requested.
```
