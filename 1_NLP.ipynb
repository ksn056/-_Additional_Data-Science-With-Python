```python
import nltk
import spacy
import pandas as pd
from textblob import TextBlob

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

**1. Tokenization**

```python
from nltk.tokenize import word_tokenize, sent_tokenize

text = "Natural Language Processing is amazing. It powers ChatGPT."

print("Word Tokenization:")
print(word_tokenize(text))

print("\nSentence Tokenization:")
print(sent_tokenize(text))
```

**2. Stemming and Lemmatization**

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ["running", "studies", "better"]

print("Stemming:")
print([stemmer.stem(w) for w in words])

print("\nLemmatization:")
print([lemmatizer.lemmatize(w) for w in words])
```

**3. Named Entity Recognition (NER)**

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Elon Musk founded SpaceX in the United States in 2002.")

print("Named Entities:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

**5. Text Preprocessing**

```python
data = pd.DataFrame({
    "text": [
        "I love NLP",
        "This is terrible",
        "Amazing product",
        "Worst experience ever"
    ]
})

data
```

**9 – Convert to Lowercase**

```python
data["text"] = data["text"].str.lower()
data
```

**Remove Stopwords**

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return ' '.join([word for word in str(text).split() if word not in stop_words])

data['cleaned'] = data['text'].apply(lambda x: remove_stopwords(x))
data
```

**TF-IDF**

**6. Word Embeddings (Convert Text to Numbers)**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data["cleaned"])

print("TF-IDF Matrix:")
print(X.toarray())
```

**7. Train a Simple Model**

```python
from sklearn.linear_model import LogisticRegression

# Labels
y = [1, 0, 1, 0]

model = LogisticRegression()
model.fit(X, y)

print("Predictions:")
print(model.predict(X))
```

**8. Build a Simple Chatbot**

```python
def chatbot(user_input):
    if "hello" in user_input.lower():
        return "Hi! How can I help you?"
    elif "nlp" in user_input.lower():
        return "NLP stands for Natural Language Processing."
    elif "bye" in user_input.lower():
        return "Goodbye!"
    else:
        return "I am still learning."

while True:
    user = input("You: ")
    if user.lower() == "exit":
        break
    print("Bot:", chatbot(user))
```

**N-GRAM**

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

text = "Natural Language Processing is a fascinating field."
tokens = word_tokenize(text)

# Generate Bigrams (N=2)
bigrams_list = list(ngrams(tokens, 2))
print("Bigrams:", bigrams_list)

# Generate Trigrams (N=3)
trigrams_list = list(ngrams(tokens, 3))
print("Trigrams:", trigrams_list)
```

**Bag of Words (BoW)**

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

corpus = [
    'I love learning NLP',
    'I love programming',
    'NLP is fascinating and I love it'
]

# Create a CountVectorizer object
vectorizer = CountVectorizer()

# Fit the vectorizer to the corpus and transform the text data
X_bow = vectorizer.fit_transform(corpus)

# Get the feature names (words in the vocabulary)
feature_names = vectorizer.get_feature_names_out()

# Convert the BoW matrix to a DataFrame for better readability
bow_df = pd.DataFrame(X_bow.toarray(), columns=feature_names)

print("Original Corpus:")
for doc in corpus:
    print(f"- {doc}")

print("\nBag of Words Representation (Counts):")
print(bow_df)
```
