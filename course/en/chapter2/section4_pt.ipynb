```python
# Importing required libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import gradio as gr

# Load pretrained model and tokenizer
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Chat history
chat_history = []

def chatbot(user_input):
    global chat_history
    # Tokenize the input and add it to the chat history
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')
    bot_input_ids = new_input_ids if len(chat_history) == 0 else torch.cat([chat_history[-1], new_input_ids], dim=-1)

    # Generate response
    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    chat_history.append(chat_history_ids)

    # Decode the response
    response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
    return response

# Create Gradio GUI
iface = gr.Interface(
    fn=chatbot,
    inputs="text",
    outputs="text",
    title="AI Chatbot for Customer Support",
    description="Type your queries to get a response from the AI chatbot."
)

# Launch the interface
iface.launch()
```

```python
# prompt: sentiment Analysis with generative AI

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def analyze_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(text)

    if scores['compound'] >= 0.05:
        sentiment = "Positive"
    elif scores['compound'] <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    return sentiment


def chatbot(user_input):
    global chat_history
    # Tokenize the input and add it to the chat history
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')
    bot_input_ids = new_input_ids if len(chat_history) == 0 else torch.cat([chat_history[-1], new_input_ids], dim=-1)

    # Generate response
    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    chat_history.append(chat_history_ids)

    # Decode the response
    response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

    # Perform sentiment analysis on the response
    sentiment = analyze_sentiment(response)

    return f"{response}\n\nSentiment: {sentiment}"


import gradio as gr
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load pretrained model and tokenizer
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Chat history
chat_history = []


# Create Gradio GUI
iface = gr.Interface(
    fn=chatbot,
    inputs="text",
    outputs="text",
    title="AI Chatbot with Sentiment Analysis",
    description="Type your queries to get a response and sentiment analysis from the AI chatbot."
)

# Launch the interface
iface.launch()
```

```python
# Importing required libraries
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import gradio as gr

# Load conversational model
chatbot_model_name = "microsoft/DialoGPT-medium"
chatbot_tokenizer = AutoTokenizer.from_pretrained(chatbot_model_name)
chatbot_model = AutoModelForCausalLM.from_pretrained(chatbot_model_name)

# Load sentiment analysis pipeline
sentiment_model_name = "cardiffnlp/twitter-roberta-base-sentiment"
sentiment_analysis = pipeline("sentiment-analysis", model=sentiment_model_name)

# Chat history
chat_history = []

# Mental support chatbot function
def mental_support_chatbot(user_input):
    global chat_history

    # Analyze sentiment
    sentiment_result = sentiment_analysis(user_input)[0]
    sentiment = sentiment_result['label']
    score = sentiment_result['score']

    # Tokenize input and generate response
    new_input_ids = chatbot_tokenizer.encode(user_input + chatbot_tokenizer.eos_token, return_tensors='pt')
    bot_input_ids = new_input_ids if len(chat_history) == 0 else torch.cat([chat_history[-1], new_input_ids], dim=-1)
    chat_history_ids = chatbot_model.generate(bot_input_ids, max_length=1000, pad_token_id=chatbot_tokenizer.eos_token_id)
    chat_history.append(chat_history_ids)
    response = chatbot_tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

    # Tailor response based on sentiment
    if sentiment == "NEGATIVE" and score > 0.6:
        response += " It seems you're going through a tough time. Remember, you're not alone. Would you like me to share some helpful resources?"
    elif sentiment == "POSITIVE" and score > 0.6:
        response += " It's great to hear you're feeling positive! Keep up the good energy!"
    else:
        response += " I'm here to listen if you'd like to share more."

    return response

# Create Gradio GUI
iface = gr.Interface(
    fn=mental_support_chatbot,
    inputs="text",
    outputs="text",
    title="AI Chatbot for Mental Support",
    description=(
        "This chatbot provides emotional support by understanding your mood and offering empathetic replies. "
        "If you're experiencing severe issues, we recommend seeking professional help."
    ),
)

# Launch the interface
iface.launch()
```
